# Ejercicio 1

```python
from functools import total_ordering

@total_ordering
class Producto:
    def __init__(self, nombre: str, precio: float, stock: int):
        self._nombre = nombre # privado
        self._precio = precio
        self._stock = stock

    def __str__(self):
        return f"Producto: {self._nombre} – Precio: ${self._precio} ({self._stock} unidades) "

    def __repr__(self): # para el desarrollador 
        return f"Producto: {self._nombre}, Precio: ${self._precio}, Stock: {self._stock}"
    
    def __eq__(self, otro):
        return self._precio == otro._precio
    
    def __lt__(self, otro):
        return self._precio < otro._precio
    
    def __add__(self, otro):
        return Producto(self._nombre, self._precio, self._stock + otro._stock)
    
    def __mul__(self, n):
        return Producto(self._nombre, self._precio * n , self._stock)
```

```python
p1 = Producto('Arroz', 1200, 50)
p2 = Producto('Arroz', 900, 30)
p3 = Producto('Leche', 1200, 20)
```

# Ejercicio 2

```python
class CuentaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    # Parte A 
    @property
    def titular(self):
        return self._titular

    @titular.setter
    def titular(self, valor):
        if len(valor) < 3:
            raise ValueError("El nombre necesita 3 caracters minimo ")
        self._titular = valor

    @property
    def saldo(self):
        return self._saldo

    @saldo.setter
    def saldo(self, valor):
        if valor < 0:
            raise ValueError("El saldo no puede ser negativo")
        self._saldo = valor

    # Parte B 
    def depositar(self, monto):
        if monto <= 0:
            raise ValueError("El monto tiene que ser mayor a 0")
        self._saldo += monto

    def retirar(self, monto):
        if monto <= 0:
            raise ValueError("El monto tiene que ser mayor a 0")
        if self._saldo - monto < 0:
            raise ValueError("Saldo insuficiente")
        self._saldo -= monto

    # Parte c
    def __str__(self):
        return f"Cuenta de {self._titular} | Saldo: ${self._saldo:}"

    def __repr__(self):
        return f"CuentaBancaria(titular='{self._titular}', saldo={self._saldo})"

    @property
    def aprobada(self):
        return self._saldo >= 10000
```

```python
cuenta = CuentaBancaria('Ana López', 50000)
#cuenta2 = CuentaBancaria('An', 232323)
```

## Parte A

```python
print(cuenta.titular)
print(cuenta.saldo)
```

## Parte B

```python
cuenta.depositar(20000)
print(cuenta.saldo) # 70000
cuenta.retirar(15000)
print(cuenta) 
```

```python
import math

class base_Figura :
    def __init__(self, color):
        self.color = color

    def area(self):
        raise NotImplementedError("Debe implementar area en la subclase")

    def perimetro(self):
        raise NotImplementedError("Debe implementar perimetro en la subclase")

    def __str__(self):
        return f"Figura {self.color} | Área: {self.area():.2f} | Perímetro: {self.perimetro():.2f}"

    def __repr__(self):
        return f"Figura(color='{self.color}')"

    def __lt__(self, other):
        if not isinstance(other, base_Figura):
            return NotImplemented
        return self.area() < other.area()
```

```python
class Circulo(base_Figura):
    def __init__(self, color, radio):
        super().__init__(color)
        if radio <= 0:
            raise ValueError("El radio debe ser mayor que 0")
        self.radio = radio

    def area(self):
        return math.pi * self.radio**2

    def perimetro(self):
        return 2 * math.pi * self.radio

    def __repr__(self):
        return f"Circulo(color='{self.color}', radio={self.radio})"
```

```python
class Rectangulo(base_Figura):
    def __init__(self, color, ancho, alto):
        super().__init__(color)
        if ancho <= 0 or alto <= 0:
            raise ValueError("Ancho y alto deben ser mayores que 0")
        self.ancho = ancho
        self.alto = alto

    def area(self):
        return self.ancho * self.alto

    def perimetro(self):
        return 2 * (self.ancho + self.alto)

    @property
    def es_cuadrado(self):
        return self.ancho == self.alto

    def __repr__(self):
        return f"Rectangulo(color='{self.color}', ancho={self.ancho}, alto={self.alto})"
```

```python
class Inventario:
    def __init__(self):
        self._productos = []

    def agregar(self, producto):
        self._productos.append(producto)

    def __len__(self):
        return len(self._productos)

    def __getitem__(self, i):
        return self._productos[i]

    def __contains__(self, producto):
        return producto in self._productos

    def __iter__(self):
        return iter(self._productos)

    def __str__(self):
        return "\n".join(str(p) for p in self._productos)

    def __repr__(self):
        return f"Inventario(n_productos={len(self)})"

    @property
    def total_valor(self):
        return sum(p.precio * p.stock for p in self._productos)

    @property
    def mas_caro(self):
        return max(self._productos, key=lambda p: p.precio)

    def buscar(self, nombre):
        nombre = nombre.lower()
        return [p for p in self._productos if nombre in p.nombre.lower()]
```

```python
class Producto:
    def __init__(self, nombre: str, precio: float, stock: int):
        self._nombre = nombre
        self._precio = precio
        self._stock = stock

    # --- PROPERTIES ---
    @property
    def nombre(self):
        return self._nombre

    @property
    def precio(self):
        return self._precio

    @property
    def stock(self):
        return self._stock

    # --- REPRESENTACIÓN ---
    def __str__(self):
        return f"Producto: {self._nombre} – Precio: ${self._precio} ({self._stock} unidades)"

    def __repr__(self):
        return f"Producto(nombre='{self._nombre}', precio={self._precio}, stock={self._stock})"

    # --- COMPARACIÓN ---
    def __eq__(self, otro):
        return self._precio == otro._precio

    def __lt__(self, otro):
        return self._precio < otro._precio

    # --- OPERACIONES ---
    def __add__(self, otro):
        return Producto(self._nombre, self._precio, self._stock + otro._stock)

    def __mul__(self, n):
        return Producto(self._nombre, self._precio * n, self._stock)
```

## Parte A

```python
inv = Inventario()
inv.agregar(Producto('Arroz', 1200, 50))
inv.agregar(Producto('Leche', 950, 30))
inv.agregar(Producto('Pan', 1800, 100))
inv.agregar(Producto('Panetón', 3500, 10))
```
