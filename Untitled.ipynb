```python
import random
from functools import reduce
```

```python
#actividad 1
numbers = [2,4,6] #lista
print(list(map(lambda x: x**2+2*x+1, numbers))) #se crea una funcion lambda que calcula la expresión: x² + 2x + 1.
```

```python
#actividad 2
velocidades = [10, 20, 30, 40, 50] #lista
print(list(map(lambda x : f"{x*1000} metros", velocidades))) #Convierte velocidades de km a metros usando map()
```

```python
#actividad 3
numbers = [2,5,8,11,14,17,20] #lista
print(list(filter(lambda x: x % 2 == 0, numbers))) #Filtra los números pares del dataset
```

```python
#actividad 4
numeross = [2,3,4,5] #lista
print(reduce(lambda x,y: x+y, numeross)) #Calcula el producto de la lista usando reduce
```

```python
#actividad 5
data = [12,15,22,35,41,8,3,55,62,90] #lista
print(sum(map(lambda x: x * 2, filter(lambda x: x >= 10, data)))) #Elimina valores menores a 10 con filter, luego Multiplicar cada valor por 2 con map y se se suma todo con sum para optimizar el codigo
```

```python
#actividad 6
data = [random.randint(0,100) for _ in range(1000000)] #lista
totsaot = list(map(lambda x: x**2, filter(lambda x: x >= 50, data))) #Filtra los valores mayores a 50, Eleva los valores al cuadrado 
print(sum(totsaot)/len(totsaot)) #Calcula el promedio final de los resultados con las funciones de sum y len
```

```python
#actividad 7-desafio avanzado
sensores = [random.uniform(10,80) for _ in range(500000)] #lista
výsledky = list(map(lambda x: x+273.15, filter(lambda x: 10 <= x <= 70, sensores))) #Filtra valores entre 10 y 70 grados celsius, y los convierte en kelvin
print(sum(výsledky)/len(výsledky)) #Calcula el promedio final de los resultados con las funciones de sum y len
```
