# ejercicio 1 "Sistema de Empleados"

```python
class Empleado:
    def __init__ (self,nombre,empresa,salario):
        self.nombre = nombre
        self.empresa = empresa
        self.salario = salario


    def calcular_salario(self):
        pass
```

```python
class GerenteVentas(Empleado):
    def __init__ (self,nombre: str,empresa: str,salario_base: int,bono_ventas: int,salario: int):
        
        super().__init__(nombre,empresa,salario)
        
        self.bono_ventas = bono_ventas
        self.salario_base = salario_base

    def calcular_salario(self):
        self.salario = self.salario_base + self.bono_ventas
        
    def descripcion(self):
        print(f"gerente de ventas: {self.nombre}|salario${self.salario}")
```

```python
class ObreroPorHora(Empleado):
    def __init__ (self,nombre: str,empresa: str,salario: int,horas_trabajadas: int,valor_hora: int):
        
        super().__init__(nombre,empresa,salario)
        
        self.horas_trabajadas = horas_trabajadas
        self.valor_hora = valor_hora
        
    def calcular_salario(self):
        self.salario = self.horas_trabajadas * self.valor_hora
        
    def descripcion(self):
        print(f"Obrero Por Hora: {self.nombre}|salario${self.salario}")
```

```python
gerente1 = GerenteVentas("anita la huerfanita", "mish",300000,123,0)
obrero1 = ObreroPorHora("anito el huerfanito", "mish",300000,123,321)
```

```python
gerente1.calcular_salario()
gerente1.descripcion()

obrero1.calcular_salario()
obrero1.descripcion()
```

# ejercicio 3 "Sistema de Figuras Geometricas(ABC)"

```python
from abc import ABC, abstractmethod
```

```python
class Figura(ABC):
    @abstractmethod
    def area(self):
        pass
        
    @abstractmethod
    def perimetro(self):
        pass

    def __init__(self,area,perimetro):
        self.area = area
        self.perimetro = perimetro
        
    def describir(self):
        return f"area:{self.area()} perimetro:{self.perimetro()}"
```

## sub clase

```python
class Circulo(Figura):

    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return 3.14159 * self.radio ** 2

    def perimetro(self):
        return 2 * 3.14159 * self.radio
```

```python
class Rectangulo(Figura):

    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def perimetro(self):
        return 2 * (self.base + self.altura)
```

```python
class Triangulo(Figura):

    def __init__(self, base, altura, hipotenusa):
        self.base = base
        self.altura = altura
        self.hipotenusa = hipotenusa

    def area(self):
        return (self.base * self.altura) / 2

    def perimetro(self):
        return self.base + self.altura + self.hipotenusa
```

## aparte

```python
def calcular_total(figuras: list):
    x = 0
    for i in figuras:
        x += i.area()
    return x
```

```python
figuras = [
    Circulo(5),
    Rectangulo(4, 6),
    Triangulo(3, 4, 5)
]

print("total", calcular_total(figuras))
for f in figuras:
    print(f.describir())
```

# ejercicio 4 "Sistema de Pagos"

```python
class MetodoPago(ABC):
    @abstractmethod
    def procesar(monto):
        pass
    @abstractmethod
    def reembolsar(monto):
        pass
```

## sub clases

```python
class PagoTarjeta(MetodoPago):
    def __init__(self, numero_tarjeta, titular):
        self.numero_tarjeta = numero_tarjeta
        self.titular = titular

    def procesar(self, monto):
        return f"Procesando ${monto} con la tarjeta de {self.titular}"

    def reembolsar(self, monto):
        return f"Reembolsando ${monto} a la tarjeta de {self.titular}"
```

```python
class PagoTransferencia(MetodoPago):
    def __init__(self, cuenta_origen, cuenta_destino):
        self.cuenta_origen = cuenta_origen
        self.cuenta_destino = cuenta_destino

    def procesar(self, monto):
        return f"Transfiriendo ${monto} de {self.cuenta_origen} a {self.cuenta_destino}"

    def reembolsar(self, monto):
        return f"Reembolsando ${monto} a {self.cuenta_origen}"
```

```python
class PagoEfectivo(MetodoPago):
    
    def procesar(self, monto):
        return f" se a procesado ${monto} en efectivo"

    def reembolsar(self, monto):
        return f"se devolvio ${monto} en efectivo"
```

## aparte

```python
def ejecutar_pago(metodo: MetodoPago, monto: float):
    return metodo.procesar(monto)

print(ejecutar_pago(PagoTarjeta("1234", "Juan"), 5000))
```
