## ejercicio 3

```python
promedio_general = df['monto_bruto'].mean()

filtro_especial = df[(df['region'].isin(['Norte', 'Sur'])) & (df['cantidad'] >= 10) & (df['monto_bruto'] > promedio_general)
]
print(f"promedio general: {promedio_general:.2f}")
print(f"registros: {len(filtro_especial)}")
print(filtro_especial.head())
```

## ejercicio 4

```python
filto_Bakan = filtro_especial['region'].value_counts()

filto_no_Bakan = df[(df['region'].isin(['Norte', 'Sur'])) & (df['monto_bruto'] > promedio_general)]['region'].value_counts()

print("Ventas con el filtro de cantidad (>= 10):")
print(filto_Bakan)
print("\nVentas sin filtro de cantidad (todas las cantidades):")
print(filto_no_Bakan)
```

## ejercicio 5

```python
def filtrar_ventas(df, region, categoria, monto_minimo):

    filtro = (df['region'] == region) & \
             (df['categoria'] == categoria) & \
             (df['monto_bruto'] >= monto_minimo)
    
    return df[filtro]

prueba1 = filtrar_ventas(df, 'Norte', 'Laptop', 1000)
prueba2 = filtrar_ventas(df, 'Sur', 'Tablet', 150)
prueba3 = filtrar_ventas(df, 'Centro', 'Accesorios', 50)

print(f"Resultado 1: {len(prueba1)} ")
print(f"Resultado 2: {len(prueba2)} ")
print(f"Resultado 3: {len(prueba3)} ")
```
