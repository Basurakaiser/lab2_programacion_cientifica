# **Ejercicio 3**

```python
dflimpio = df.copy()

dflimpio["temp_max"] = dflimpio["temp_max"].fillna(dflimpio["temp_max"].mean())
dflimpio["temp_min"] = dflimpio["temp_min"].fillna(dflimpio["temp_min"].mean())

dflimpio["humedad_pct"] = dflimpio["humedad_pct"].fillna(dflimpio["humedad_pct"].median())
dflimpio["precipitacion_mm"] = dflimpio["precipitacion_mm"].fillna(0.0)

filas_antes = df.shape[0]
dflimpio = dflimpio.dropna(subset=["condicion"])

filas_despues = dflimpio.shape[0]
print("Filas eliminadas:", filas_antes - filas_despues)
assert dflimpio.isnull().sum().sum() == 0
```

```python
df.dropna().shape
```

### P3.1 ¿Por qué se usó fillna(mean()) para temp_max y fillna(median()) para humedad_pct?

#### *por que la media representa bien a la temperatura al ser mas estable, en cambio la mediana es mejor para valores que pueden ser extremos como en la humedad*

### P3.2 ¿Cuántas filas se eliminaron al aplicar dropna(subset=['condicion'])? ¿Es aceptable esta pérdida? 

#### *2 filas, es aceptable la perdida ya que es baja respecto al total del data set*

### P3.3 ¿Qué sucede si se aplica df.dropna() sin parámetros? ¿Cuántas filas quedarían? Pruébalo.

#### *elimina las filas que tienen un dato nulo, quedando 16 filas*
